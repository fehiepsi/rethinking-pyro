#### Code 5.1

```python
# load data
d = pd.read_csv("Data/WaffleDivorce.csv", sep=";")

# standardize predictor
d['MedianAgeMarriage_s'] = ((d.MedianAgeMarriage - d.MedianAgeMarriage.mean()) /
                            d.MedianAgeMarriage.std())

# fit model
def m5_1(age, divorce):
    a = pyro.sample("a", dist.normal, ng_ones(1)*10, ng_ones(1)*10)
    bA = pyro.sample("bA", dist.normal, ng_zeros(1), ng_ones(1))
    mu = a + bA * age
    sigma = pyro.sample("sigma", dist.uniform, ng_zeros(1), ng_ones(1)*10)
    return pyro.sample("divorce", dist.normal, mu, sigma.expand_as(mu), obs=divorce)

age = Variable(torch.Tensor(d.MedianAgeMarriage_s))
divorce = Variable(torch.Tensor(d.Divorce))
m5_1_qa = Laplace(m5_1, [("a", 10), ("bA", 0), ("sigma", 5)]).fit(
    age, divorce, optim_params={"lr": 0.1})
```

#### Code 5.2

```python
# compute percentile interval of mean
MAM_seq = np.linspace(-3, 3.5, num=30)
post = m5_1_qa.extract_samples()
mu = post[:,0].reshape(-1, 1) + np.outer(post[:,1], MAM_seq)
mu_PI = np.apply_along_axis(PI, 0, mu).T

# plot it all
plt.plot(d.MedianAgeMarriage_s, d.Divorce, 'C0o')
plt.plot(MAM_seq, mu.mean(0), 'C2')
plt.fill_between(MAM_seq, mu_PI[:,0], mu_PI[:,1], color='C2', alpha=0.25)
plt.xlabel('Meadian Age Marriage', fontsize=14)
plt.ylabel('Divorce', fontsize=14);
```

#### Code 5.3

```python
d["Marriage_s"] = (d.Marriage - d.Marriage.mean()) / d.Marriage.std()

def m5_2(marriage, divorce):
    a = pyro.sample("a", dist.normal, ng_ones(1)*10, ng_ones(1)*10)
    bR = pyro.sample("bR", dist.normal, ng_zeros(1), ng_ones(1))
    mu = a + bR * marriage
    sigma = pyro.sample("sigma", dist.uniform, ng_zeros(1), ng_ones(1)*10)
    return pyro.sample("divorce", dist.normal, mu, sigma.expand_as(mu), obs=divorce)

marriage = Variable(torch.Tensor(d.Marriage_s))
m5_2_qa = Laplace(m5_2, [("a", 10), ("bR", 0), ("sigma", 5)]).fit(
    age, divorce, optim_params={"lr": 0.1})
```

#### Code 5.4

```python
def m5_3(marriage, age, divorce):
    a = pyro.sample("a", dist.normal, ng_ones(1)*10, ng_ones(1)*10)
    bR = pyro.sample("bR", dist.normal, ng_zeros(1), ng_ones(1))
    bA = pyro.sample("bA", dist.normal, ng_zeros(1), ng_ones(1))
    mu = a + bR * marriage + bA * age
    sigma = pyro.sample("sigma", dist.uniform, ng_zeros(1), ng_ones(1)*10)
    return pyro.sample("divorce", dist.normal, mu, sigma.expand_as(mu), obs=divorce)

m5_3_qa = Laplace(m5_3, [("a", 10), ("bR", 0), ("bA", 0), ("sigma", 5)]).fit(
    marriage, age, divorce, optim_params={"lr": 0.1})
m5_3_qa.precis()
```

```python
def simcoll(r = 0.9):
    milk['x'] = stats.norm.rvs(size=len(milk), loc = r * milk['perc.fat'],
                       scale = np.sqrt((1 - r**2) * milk['perc.fat'].var()))
    X = np.column_stack((milk['perc.fat'], milk['x']))
    m = smf.OLS(milk['kcal.per.g'], X).fit()
    cov = m.cov_params()
    return (np.diag(cov)[1])**0.5
    
def repsimcoll(r= 0.9, N = 100):
    stddev = [simcoll(r) for _ in range(N)]
    return np.mean(stddev)

lista = []
for i in np.arange(start = 0, stop = 0.99, step = 0.01):   
    lista.append(repsimcoll (r= i, N = 100))

plt.plot(np.arange(start = 0, stop = 0.99, step = 0.01), lista)
plt.xlabel('correlation', fontsize=14)
plt.ylabel('stddev', fontsize=14);
```
